```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
import re
import urllib
import requests
from time import sleep
from selenium.common.exceptions import NoSuchElementException
```

```python
options = webdriver.ChromeOptions()
options.add_argument('headless') 
url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=103#&date=%2000:00:00&page=1'
path = 'C:/tool/chromedriver.exe'
driver = webdriver.Chrome(path,options=options)
driver.get(url)
```

```python
politics = []
economy = []
world = []
science = []

for cate in [100,101,105,104]:
    for page in range(1,31):
        url = f'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1={cate}#&date=%2000:00:00&page={page}'
        driver.get(url)
        print('현재위치: ', driver.current_url)
        sleep(0.7)
        for i in range(1,5):
            sleep(0.5)
            for j in range(1,6):
                driver.find_element_by_xpath(f'//*[@id="section_body"]/ul[{i}]/li[{j}]/dl/dt[2]/a').send_keys(Keys.ENTER)
                # 기사 제목 클릭하여 이동
                sleep(0.5)
                article = driver.find_element_by_xpath('//*[@id="articleBodyContents"]') # 이동한 기사내용
                a = re.sub('[^a-zA-Z가-힣\s]+','',article.text) # 영어, 대소문자, 한글, 스페이스바를 제외한 모든걸 공백으로 바꿔준다.
                b = re.sub('[\n]+',' ',a) # 줄바꿈을 스페이바로 치환
                if cate == 100: politics.append(b)
                elif cate == 101: economy.append(b)
                elif cate == 105: science.append(b)
                elif cate == 104: world.append(b)
                driver.back()
                sleep(0.5)
driver.quit()
```
